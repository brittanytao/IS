```python
from __future__ import print_function
import urllib3 
import unicodedata
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
import datetime
import requests
warnings.filterwarnings('ignore')
```

```python
#The Button: A Nuclear Fable
url='https://www.newyorker.com/magazine/2018/01/29/the-button-a-nuclear-fable'
```

```python
http = urllib3.PoolManager()
response = http.request('GET',url)
response.status
#We find out weather the url is valid for scrapping. If we get 200, it is valid.
```

```python
s=BeautifulSoup(response.data, "lxml")
#Inspects the site and stores all the value to s.
```

```python
s.find_all('a')
#Finds the html content for the url.
```

```python
for a in s.find_all('a', href = True):
    print("url:", a['href'],", Description:",a.text)
#Extracts the urls from the html content and prints it with its description.
```

```python
links=[]
for url in s.find_all('a', href = True):
    if url ['href'].startswith('http') and 'newyorker' not in url['href']:
        tmp=[]
        tmp.append(url['href'])
        tmp.append(url.text)
        tmp.append(datetime.datetime.now().strftime("%A %B %d, %Y %I:%M %p"))
        try:
            response = http.request('GET',url['href'])
            tmp.append(response.status)
        except:
            pass
        print (tmp)
        links.append(tmp)
#We save the url, the description, the date-time, and response status in a temperary list, the move it to another list.
```

```python
import csv
with open("output.csv",'w') as resultFile:
    wr = csv.writer(resultFile, dialect='excel')
    wr.writerows(links)
#Export to csv named output
```
